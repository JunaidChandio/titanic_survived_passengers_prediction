```python
def imput_age(cols):
  Age=cols[0]
  Pclass=cols[1]

  if pd.isnull(Age):
    if Pclass==1:
      return 37
    elif Pclass==2:
      return 29
    else:
      return 24

  else:
    return Age
```

```python
df['Age']=df[['Age','Pclass']].apply(imput_age,axis=1)
```

```python
df.drop('Cabin',axis=1,inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df.head()
```

```python
df.drop(['Name','Ticket'],axis=1,inplace=True)
```

```python
object=['Sex','Embarked']
```

```python
for colname in object:
  df[colname]=df[colname].astype('category')
```

```python
df
```

```python
df.describe()
```

```python
df.drop('PassengerId',axis=1,inplace=True)
```

```python
df.shape
```

```python
df.select_dtypes(['category']).columns
```

```python
sex=pd.get_dummies(df['Sex'],drop_first=True)
Embarked=pd.get_dummies(df['Embarked'],drop_first=True)
```

```python
df.head()
```

```python
df=pd.concat([df,sex,Embarked],axis=1)
df.head(10)
```

```python
df.drop(['Sex','Embarked'],axis=1,inplace=True)
```

```python
df.head()
```

```python
x=df.drop('Survived',axis=1)
y=df['Survived']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lreg=LogisticRegression()
```

```python
lreg.fit(X_train,y_train)
```

```python
y_pred=lreg.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import classification_report,accuracy_score
```

```python
lreg_acccuracy=round(accuracy_score(y_test,y_pred)*100,2)
```

```python
print("Accuracy",lreg_acccuracy,"%")
```

```python
probalilities=lreg.predict_log_proba(X_test)[:,1]
print(probalilities.shape)
print(probalilities)
```

```python
for i in range(len(y_test)):
  print(round(probalilities[i],3),"-->",y_pred[i])
```

```python
from sklearn.tree import DecisionTreeClassifier
DR=DecisionTreeClassifier()
DR.fit(X_train,y_train)
```

```python
dtree_predict=DR.predict(X_test)
```

```python
Dtree_accuracy=round(accuracy_score(y_test,dtree_predict)*100,2)
print("decision tree accuracy",Dtree_accuracy)
```

```python
from sklearn.ensemble import RandomForestClassifier
RF=RandomForestClassifier()
RF.fit(X_train,y_train)
```

```python
RF_predict=RF.predict(X_test)
```

```python
RF_accuracy=round(accuracy_score(y_test,RF_predict)*100,2)
print("Randomforest accuracy",RF_accuracy)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
GBR=GradientBoostingClassifier()
GBR.fit(X_train,y_train)
```

```python
GBR_predict=GBR.predict(X_test)
```

```python
GBR_accuracy=round(accuracy_score(y_test,GBR_predict)*100,2)
print("GradientBoosting Accuracy",GBR_accuracy)
```
